# DATA PREPROCESSING

```python
df = df.drop('Unnamed: 0', axis =1)
df = df.drop('scout_id', axis=1)
```

```python
df = df[df['rating_num'] != 0]
```

```python
test_players = df[df['rating_num'] <= 6].sample(n=1000, random_state=42)

# drop the corresponding rows from the original dataframe
df = df.drop(test_players.index)
```

```python
test_players
```

```python
test_players['rating_num'].value_counts()
```

```python
test_player_targets = test_players.iloc[:,-1]
test_players_features = test_players.drop('rating_num', axis = 1)


test_player_targets
```

# K Nearest Neighbour Regressor

```python
from sklearn.model_selection import train_test_split
y = df['rating_num']
X = df.drop(['rating_num'], axis = 1)
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
# initialize the KNN regressor with k=5
knn = KNeighborsRegressor(n_neighbors=5)

# fit the model to the data
knn.fit(X, y)

# predict the ratings of the test data
y_pred = knn.predict(test_players_features)

# evaluate the model performance using mean squared error
mse = mean_squared_error(test_player_targets, y_pred)
print('Mean squared error:', mse)


# Create a dataframe to store the player id, original rating, and new rating for the test players
test_player_ratings = pd.DataFrame(index=test_players.index)
test_player_ratings['player_id'] = test_player_targets.index
test_player_ratings['original_rating'] = test_player_targets
test_player_ratings['new_rating'] = y_pred

# Display the test player ratings dataframe
print(test_player_ratings)
```

# TEST PLAYER NEW RATING

```python
test_player_ratings
```

```python
# Add a new column to the test player ratings dataframe for the difference between the actual rating and predicted rating
test_player_ratings['rating_diff'] = abs(test_player_ratings['original_rating'] - test_player_ratings['new_rating'])

# Sort the test player ratings dataframe by the rating difference in descending order
sorted_player_ratings = test_player_ratings.sort_values('rating_diff', ascending=False)

# Display the sorted player ratings dataframe
print(sorted_player_ratings)
```
